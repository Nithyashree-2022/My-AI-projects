NORMALIZATION

```python
cols_to_norm=['user_timezone','tweet_id','tweet_count','sidebar_color','profileimage',
     'link_color','fav_number','created','_last_judgment_at',
     ]
```

```python
#mention all numeric columns

user_timezone=tf.feature_column.numeric_column('user_timezone')
tweet_id=tf.feature_column.numeric_column('tweet_id')
tweet_count=tf.feature_column.numeric_column('tweet_count')
sidebar_color=tf.feature_column.numeric_column('sidebar_color')
profileimage=tf.feature_column.numeric_column('profileimage')
link_color=tf.feature_column.numeric_column('link_color')
fav_number=tf.feature_column.numeric_column('fav_number')
created=tf.feature_column.numeric_column('created')
profile_yn=tf.feature_column.numeric_column('profile_yn')
_last_judgment_at=tf.feature_column.numeric_column('_last_judgment_at')
_unit_state=tf.feature_column.numeric_column('_unit_state')

```

```python
df.info()
```

```python
df[cols_to_norm] = df[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
df.head()
```

```python
x=df[['user_timezone','tweet_id','tweet_count','sidebar_color','profileimage',
     'link_color','fav_number','created','profile_yn','_last_judgment_at',
     '_unit_state']]
y=df['gender2']
```

```python
df['gender2'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=102)
```

# DNN Classifier using Tensorflow

```python
features=[user_timezone,tweet_id,tweet_count,sidebar_color,profileimage,
    link_color,fav_number,created,profile_yn,_last_judgment_at,
     _unit_state]
```

```python
#create DNNClassifier model:

model=tf.estimator.DNNClassifier(hidden_units=[30,30,30],feature_columns=features,
                                 n_classes=4,
                                 activation_fn=tf.nn.softmax
                                ,optimizer=tf.train.AdamOptimizer(learning_rate=0.01))
```

```python
#create train i/p func:

train_ip_func=tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
#train the DNN Classifier model:

model.train(input_fn=train_ip_func,steps=1000)
```

```python
#input function to test our model:

test_ip_func=tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
#test the dnn classifier model:

model.evaluate(test_ip_func)
```
